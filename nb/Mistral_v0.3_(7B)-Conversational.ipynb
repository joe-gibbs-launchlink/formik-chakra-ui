### Data Prep
We now use the `ChatML` format for conversation style finetunes. We use [Open Assistant conversations](https://huggingface.co/datasets/philschmid/guanaco-sharegpt-style) in ShareGPT style. ChatML renders multi turn conversations like below:

```
<|im_start|>system
You are a helpful assistant.<|im_end|>
<|im_start|>user
What's the capital of France?<|im_end|>
<|im_start|>assistant
Paris.
```

**[NOTE]** To train only on completions (ignoring the user's input) read TRL's docs [here](https://huggingface.co/docs/trl/sft_trainer#train-on-completions-only).

We use our `get_chat_template` function to get the correct chat template. We support `zephyr, chatml, mistral, llama, alpaca, vicuna, vicuna_old` and our own optimized `unsloth` template.

Normally one has to train `<|im_start|>` and `<|im_end|>`. We instead map `<|im_end|>` to be the EOS token, and leave `<|im_start|>` as is. This requires no additional training of additional tokens.

Note ShareGPT uses `{"from": "human", "value" : "Hi"}` and not `{"role": "user", "content" : "Hi"}`, so we use `mapping` to map it.

For text completions like novel writing, try this [notebook](https://colab.research.google.com/github/unslothai/notebooks/blob/main/nb/Mistral_(7B)-Text_Completion.ipynb).

````python
import json
from xml.sax.saxutils import escape
from datasets import Dataset # To create a Hugging Face Dataset object

# --- XML Conversion Functions ---
def _json_to_xml_recursive(data, parent_key_name=None):
    """
    Recursively converts a Python dictionary or list to an XML string fragment.
    Handles basic types, nested dictionaries, and lists.
    Null values are represented as self-closing tags.
    """
    xml_parts = []
    if isinstance(data, dict):
        for key, value in data.items():
            safe_key = escape(str(key)) # Ensure key is a string and escaped
            if value is None:
                xml_parts.append(f"<{safe_key} />") # Self-closing tag for None
            elif isinstance(value, list):
                xml_parts.append(f"<{safe_key}>")
                for item in value:
                    item_tag = "item" # Default tag for list items
                    # Special handling for 'core_message_and_key_facts' list items
                    if key == "core_message_and_key_facts":
                        item_tag = "fact"
                    # Add other specific list item tag names here if needed for other array fields

                    xml_parts.append(f"<{escape(item_tag)}>")
                    if isinstance(item, (dict, list)): # Nested structure in list
                        xml_parts.append(_json_to_xml_recursive(item, item_tag))
                    else:
                        xml_parts.append(escape(str(item))) # Simple value in list
                    xml_parts.append(f"</{escape(item_tag)}>")
                xml_parts.append(f"</{safe_key}>")
            elif isinstance(value, dict): # Nested dictionary
                xml_parts.append(f"<{safe_key}>")
                xml_parts.append(_json_to_xml_recursive(value, key))
                xml_parts.append(f"</{safe_key}>")
            else: # Simple value
                xml_parts.append(f"<{safe_key}>{escape(str(value))}</{safe_key}>")
    elif isinstance(data, list): # If the data itself is a list (e.g., a list of objects not under a key)
        for item_idx, item in enumerate(data):
            item_tag = parent_key_name + "_item" if parent_key_name else f"item_{item_idx}"
            xml_parts.append(f"<{escape(item_tag)}>")
            if isinstance(item, (dict, list)):
                 xml_parts.append(_json_to_xml_recursive(item, item_tag))
            else:
                xml_parts.append(escape(str(item)))
            xml_parts.append(f"</{escape(item_tag)}>")

    return "".join(xml_parts)

def convert_json_to_xml(json_data_dict, root_name="extracted_data"):
    """
    Converts a Python dictionary (from JSON) to a complete XML string
    with a specified root element name.
    """
    if not isinstance(json_data_dict, dict):
        print(f"Warning: Input 'json_data_dict' for XML conversion was not a dictionary (type: {type(json_data_dict)}). Returning error XML.")
        return f"<{escape(root_name)}><error>Invalid or empty input data for XML conversion</error></{escape(root_name)}>"

    xml_content = _json_to_xml_recursive(json_data_dict)
    return f"<{escape(root_name)}>{xml_content}</{escape(root_name)}>"

# --- Data Loading and Transformation Function ---
def load_and_prepare_synthetic_dataset(jsonl_file_path="synthetic_outputs.jsonl"):
    """
    Loads data from a .jsonl file, removes specified metadata keys from
    'extracted_data', converts 'extracted_data' to XML,
    and formats it into the ShareGPT-style conversation list.
    """
    processed_conversations = []
    keys_to_remove = [
        "source_document_name",
        "extraction_timestamp", # Assuming this is metadata you want to remove from the final XML
        "original_base_json_input_name",
        "generation_method",
        "generation_timestamp",
        "source_document_length_chars" # This was in the original extracted_data but often not needed for the model input
    ]

    try:
        with open(jsonl_file_path, 'r', encoding='utf-8') as f:
            for i, line in enumerate(f):
                try:
                    if not line.strip():
                        print(f"Skipping empty line at index {i}.")
                        continue
                    data_entry = json.loads(line.strip())

                    extracted_data_json = data_entry.get("extracted_data")
                    output_markdown = data_entry.get("output_markdown_content")

                    if extracted_data_json is None:
                        print(f"Warning: 'extracted_data' is missing or null in line {i+1}. Creating error XML.")
                        xml_input = convert_json_to_xml({}, "extracted_data")
                    else:
                        # Make a copy to modify, to avoid altering the original dict if it's referenced elsewhere
                        modified_extracted_data = extracted_data_json.copy()
                        for key in keys_to_remove:
                            if key in modified_extracted_data:
                                del modified_extracted_data[key]
                        xml_input = convert_json_to_xml(modified_extracted_data, "extracted_data")

                    if output_markdown is None:
                        print(f"Warning: 'output_markdown_content' is missing or null in line {i+1}. Using empty string.")
                        output_markdown = ""

                    conversation = [
                        {"from": "human", "value": xml_input},
                        {"from": "gpt", "value": output_markdown}
                    ]
                    processed_conversations.append(conversation)

                except json.JSONDecodeError:
                    print(f"Skipping line {i+1} due to JSONDecodeError: {line.strip()[:100]}...")
                    continue
                except Exception as e:
                    print(f"Error processing line {i+1}: {e}. Line content: {line.strip()[:100]}...")
                    continue

    except FileNotFoundError:
        print(f"Error: The file '{jsonl_file_path}' was not found.")
        return Dataset.from_dict({"conversations": []})
    except Exception as e:
        print(f"An unexpected error occurred while trying to read '{jsonl_file_path}': {e}")
        return Dataset.from_dict({"conversations": []})

    if not processed_conversations:
        print("Warning: No valid conversations were processed. Returning an empty dataset.")

    return Dataset.from_dict({"conversations": processed_conversations})

# --- Example Usage (integrate this into your Colab notebook) ---

# 1. Make sure 'synthetic_outputs.jsonl' is accessible in your Colab environment.
jsonl_file = "synthetic_outputs.jsonl"

# 2. Load and prepare your synthetic dataset
print(f"Loading and preparing dataset from '{jsonl_file}'...")
synthetic_dataset = load_and_prepare_synthetic_dataset(jsonl_file)

print(f"\nSuccessfully loaded and processed {len(synthetic_dataset)} entries.")

if len(synthetic_dataset) > 0:
    print("\nExample of one processed entry (ShareGPT style, before applying chat template):")
    print(json.dumps(synthetic_dataset[0]['conversations'], indent=2))

    # ... (rest of your tokenizer setup and dataset mapping code) ...

else:
    print("\nNo data was loaded. Please check your 'synthetic_outputs.jsonl' file and path.")
````

If you're looking to make your own chat template, that also is possible! You must use the Jinja templating regime. We provide our own stripped down version of the `Unsloth template` which we find to be more efficient, and leverages ChatML, Zephyr and Alpaca styles.

More info on chat templates on [our wiki page!](https://github.com/unslothai/unsloth/wiki#chat-templates)

````python
from unsloth.chat_templates import get_chat_template

# Assume 'tokenizer' is already loaded and initialized, for example:
# from transformers import AutoTokenizer
# tokenizer = AutoTokenizer.from_pretrained("your-base-model-name")

# 1. Define your custom template and EOS token
unsloth_template = \
    "{{ bos_token }}"\
    "{{ 'You are a helpful assistant to the user\\n' }}"\
    "{% for message in messages %}"\
        "{% if message['role'] == 'user' %}"\
            "{{ '>>> User: ' + message['content'] + '\\n' }}"\
        "{% elif message['role'] == 'assistant' %}"\
            "{{ '>>> Assistant: ' + message['content'] + eos_token + '\\n' }}"\
        "{% endif %}"\
    "{% endfor %}"\
    "{% if add_generation_prompt %}"\
        "{{ '>>> Assistant: ' }}"\
    "{% endif %}"

unsloth_eos_token = "eos_token" # Or "</s>" if that's what your model/Unsloth expects for its true EOS

# 2. Apply the chat template to your tokenizer
# The 'if False:' is removed so this code executes.
tokenizer = get_chat_template(
    tokenizer,
    chat_template = (unsloth_template, unsloth_eos_token,), # Provide as a tuple
    mapping = {"role" : "from", "content" : "value", "user" : "human", "assistant" : "gpt"}, # For ShareGPT style
    map_eos_token = True, # Important for Unsloth to handle EOS correctly, often maps to </s>
)

def formatting_prompts_func(examples):
  # 'examples' will be a batch from your Hugging Face dataset.
  # examples["conversations"] will be a list of conversations.
  # Each 'convo' in 'convos' is a list of turns (dictionaries).
  convos = examples["conversations"]

  texts = [tokenizer.apply_chat_template(convo, tokenize=False, add_generation_prompt=False) for convo in convos]

  # Returns a dictionary mapping the new column name "text" to the list of processed strings.
  return {"text": texts}

print("Formatting dataset...")
# Apply the formatting function to create the "text" column
# Ensure synthetic_dataset is not empty before mapping
if len(synthetic_dataset) > 0:
    dataset_ready_for_trainer = synthetic_dataset.map(
        formatting_prompts_func,
        batched=True,  # Process data in batches for efficiency
        # num_proc=2,  # Optional: Use multiple processes if your dataset is large
        remove_columns=["conversations"] # Optional: Remove the original column if no longer needed
    )
    print("Dataset formatted.")
    print(f"First example text: {dataset_ready_for_trainer[0]['text']}")
else:
    print("Synthetic dataset is empty. Cannot format.")
    # Handle this case appropriately, perhaps by exiting or raising an error
    # For now, creating an empty shell to avoid subsequent errors if trainer is still called.
    from datasets import Dataset
    dataset_ready_for_trainer = Dataset.from_dict({"text":[]})

# Now your 'tokenizer' is updated with the custom chat template.
# You can proceed to use it with your dataset and the formatting_prompts_func:

# Example (assuming synthetic_dataset is loaded as per previous steps):
# def formatting_prompts_func(examples):
#     convos = examples["conversations"]
#     texts = [tokenizer.apply_chat_template(convo, tokenize = False, add_generation_prompt = False) for convo in convos]
#     return { "text" : texts, }
#
# if len(synthetic_dataset) > 0:
#     try:
#         dataset_formatted = synthetic_dataset.map(formatting_prompts_func, batched = True)
#         print("\nExample of one formatted text entry (after applying custom chat template):")
#         print(dataset_formatted[0]['text'])
#     except Exception as e:
#         print(f"\nAn error occurred during the .map() operation: {e}")
# else:
#     print("\nDataset is empty, cannot apply formatting.")
````

### Train the model
Now let's use Huggingface TRL's `SFTTrainer`! More docs here: [TRL SFT docs](https://huggingface.co/docs/trl/sft_trainer). We do 60 steps to speed things up, but you can set `num_train_epochs=1` for a full run, and turn off `max_steps=None`. We also support TRL's `DPOTrainer`!

````python
from trl import SFTTrainer
from transformers import TrainingArguments
from unsloth import is_bfloat16_supported
# Ensure 'model', 'tokenizer', 'max_seq_length' are defined
# Use 'dataset_ready_for_trainer' which now has the "text" column
if len(dataset_ready_for_trainer) > 0 : # Only proceed if dataset has content
    trainer = SFTTrainer(
        model=model,
        tokenizer=tokenizer,
        train_dataset=dataset_ready_for_trainer, # Pass the pre-processed dataset
        # formatting_func=None, # REMOVE or set to None
        dataset_text_field="text",      # Tell SFTTrainer to use the "text" column directly
        max_seq_length=max_seq_length,
        dataset_num_proc=2,
        packing=False,
        args=TrainingArguments(
            per_device_train_batch_size=2,
            gradient_accumulation_steps=4,
            warmup_steps=5,
            max_steps=60,
            learning_rate=2e-4,
            fp16=not is_bfloat16_supported(),
            bf16=is_bfloat16_supported(),
            logging_steps=1,
            optim="adamw_8bit",
            weight_decay=0.01,
            lr_scheduler_type="linear",
            seed=3407,
            output_dir="outputs",
            report_to="none",
        ),
    )
    print("SFTTrainer initialized successfully with pre-formatted dataset.")
else:
    print("Cannot initialize SFTTrainer: Formatted dataset is empty.")
````

### Inference
Let's run the model! Since we're using `ChatML`, use `apply_chat_template` with `add_generation_prompt` set to `True` for inference.

````python
from unsloth.chat_templates import get_chat_template

tokenizer = get_chat_template(
    tokenizer,
    chat_template = "chatml", # Supports zephyr, chatml, mistral, llama, alpaca, vicuna, vicuna_old, unsloth
    mapping = {"role" : "from", "content" : "value", "user" : "human", "assistant" : "gpt"}, # ShareGPT style
    map_eos_token = True, # Maps <|im_end|> to </s> instead
)

FastLanguageModel.for_inference(model) # Enable native 2x faster inference

messages = [
    {"from": "human", "value": "Continue the fibonnaci sequence: 1, 1, 2, 3, 5, 8,"},
]
inputs = tokenizer.apply_chat_template(
    messages,
    tokenize = True,
    add_generation_prompt = True, # Must add for generation
    return_tensors = "pt",
).to("cuda")

outputs = model.generate(input_ids = inputs, max_new_tokens = 64, use_cache = True)
tokenizer.batch_decode(outputs)
````

### GGUF / llama.cpp Conversion
To save to `GGUF` / `llama.cpp`, we support it natively now! We clone `llama.cpp` and we default save it to `q8_0`. We allow all methods like `q4_k_m`. Use `save_pretrained_gguf` for local saving and `push_to_hub_gguf` for uploading to HF.

Some supported quant methods (full list on our [Wiki page](https://github.com/unslothai/unsloth/wiki#gguf-quantization-options)):
* `q8_0` - Fast conversion. High resource use, but generally acceptable.
* `q4_k_m` - Recommended. Uses Q6_K for half of the attention.wv and feed_forward.w2 tensors, else Q4_K.
* `q5_k_m` - Recommended. Uses Q6_K for half of the attention.wv and feed_forward.w2 tensors, else Q5_K.

````python
# Save to 8bit Q8_0
if True: model.save_pretrained_gguf("model", tokenizer,)
if False: model.push_to_hub_gguf("hf/model", tokenizer, token = "")

# Save to 16bit GGUF
if False: model.save_pretrained_gguf("model", tokenizer, quantization_method = "f16")
if False: model.push_to_hub_gguf("hf/model", tokenizer, quantization_method = "f16", token = "")

# Save to q4_k_m GGUF
if False: model.save_pretrained_gguf("model", tokenizer, quantization_method = "q4_k_m")
if False: model.push_to_hub_gguf("hf/model", tokenizer, quantization_method = "q4_k_m", token = "")
````
